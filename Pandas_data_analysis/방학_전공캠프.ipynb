## 예제 1-1 딕셔너리 / 시리즈 변환
#### __딕셔너리 -> 시리즈 변환 : ```pandas.Series(딕셔너리)```__

````python
# pandas 불러오기 
import pandas as pd

# k:v 구조를 갖는 딕셔너리를 만들고, 변수 dict_data에 저장
dict_data = {'a': 1, 'b': 2, 'c': 3}

# 판다스 Series() 함수로 딕셔너리(dict_data)를 시리즈로 변환. 변수 sr에 저장 
sr = pd.Series(dict_data)

# 변수 sr의 자료형 출력
print(type(sr))
print('\n')

# 변수 sr에 저장되어 있는 시리즈 객체를 출력
print(sr)
````

## 예제 1-2 시리즈 인덱스
#### __인덱스 배열 : ```Series객체.index```__   
#### __데이터 값 배열 : ```Series객체.values```__

````python
# 리스트를 시리즈로 변환하여 변수 sr에 저장
list_data = ['2019-01-02', 3.14, 'ABC', 100, True]
sr = pd.Series(list_data)
print(sr)

# 인덱스 배열은 변수 idx에 저장. 데이터 값 배열은 변수 val에 저장
idx = sr.index
val = sr.values
print(idx)
print('\n')
print(val)
````

## 예제 1-3 시리즈 원소 선택

````python
# 튜플을 시리즈로 변환(인덱스 옵션 지정)
tup_data = ('영인', '2010-05-01', '여', True)
sr = pd.Series(tup_data, index=['이름', '생년월일', '성별', '학생여부'])
print(sr)
````

````python
# 원소 1개 선택
print(sr[0])        # sr의 1번째 원소를 선택(정수형 위치 인덱스)
print(sr['이름'])   # '이름' 라벨을 가진 원소를 선택(인덱스 이름)
````

````python
# 여러개의 원소를 선택(인덱스 리스트 활용)
print(sr[[1, 2]])
print('\n')
print(sr[['생년월일', '성별']])
````

````python
# 여러 원소 선택(인덱스 범위 지정)
print(sr[1:2])
print('\n')
print(sr['생년월일' : '성별'])
````

## 예제 1-4 딕셔너리
#### __딕셔너리 -> 데이터 프레임 변혼 : ```pandas.DataFrame(딕셔너리 객체)```__

````python
# 열 이름으로 key로 하고, 리스트를 value로 갖는 딕셔너리 정의
dict_data = {'c0' : [1,2,3], 'c1' : [4,5,6], 'c2' : [7,8,9], 'c3': [10,11,12], 'c4': [13,14,15]}

# 판다스 DataFrame() 함수로 딕셔너리를 데이터프레임으로 변환. 변수 df에 저장
df = pd.DataFrame(dict_data)

# df의 자료형 출력
print(type(df))
print('\n')
# 변수 df에 저장되어 있는 데이터프레임 객체를 출력
print(df)
df
````

## 예제 1-5 행 인덱스 / 열 이름 설정
#### __행 인덱스/ 열 이름 설정 : ```pandas.DataFrame(2차원 배열, index = 행 인덱스 배열, columns=열 이름 배열)```__
#### __행 인덱스 변경 : ```DataFrame 객체.index = 새로운 행 인덱스 배열```__
#### __열 이름 변경 : ```DataFrame 객체.columns = 새로운 열 이름 배열```__

````python
# 행 인덱스 , 열 이름 지정 DF 만들기
df = pd.DataFrame([[15, '남', '덕영중'], [17, '여', '수리중']], index = ['준서', '예은'], columns=['나이', '성별', '학교'])

# 행 인덱스, 열 이름 확인
print(df)  # DF 출력
print('\n')
print(df.index)  # 행 인덱스 출력
print('\n')
print(df.columns)  # 열 이름 출력
````

````python
# 행 인덱스, 열 이름 변경
df.index=['학생1', '학생2']
df.columns=['연령', '남녀', '소속']
print(df)      # 데이터프레임
print('\n')
print(df.index)      # 행 인덱스
print('\n')
print(df.columns)    # 열 이름
````

## 예제 1-6 행 인덱스/ 열 이름 변경
#### __행 인덱스 변경 : ```DataFrame 객체.rename(index={기존 인덱스 : 새 인덱스,...})```__
#### __열 이름 변경 : ```DataFrame 객체.rename(columns={기존 이름 : 새이름, ....})```__

````python
import pandas as pd

# 행 인덱스/ 열 이름 지정하여 데이터 프레임 만들기
df = pd.DataFrame([[15, '남', '덕영중'], [17, '여', '수리중']], index=['준서', '예은'], columns=['나이', '성별', '학교'])

# 데이터프레임 df 출력
print(df)
print('\n')

# 열 이름 중 '나이'를 '연령'으로 '성벌'을 '남녀'로 '학교'를 '소속'으로 변경
df.rename(columns={'나이' : '연령', '성별':'남녀', '학교':'소속'})

# df의 행 인덱스 중에서 '준서'를 '학생1'로 '예은'을 '학생2'로 변경
df.rename(index={'준서':'학생1', '예은':'학생2'}, inplace=True)

# df 출력
print(df)
````

## 예제 1-7 행 삭제
#### __행 삭제 : ```DataFrame 객체.drop(행 인덱스 또는 배열, axis=0)```__
#### __열 삭제 : ```DataFrame 객체.drop(열 이름 또는 배열, axis=1)```__

````python
# DataFrame() 함수로 데이터프레임 변환
exam_data = {'수학' : [90, 80, 70], '영어' : [98, 89, 85], '음악' : [85, 95, 100], '체육' : [100, 90, 90]}
df = pd.DataFrame(exam_data, index=['서준', '우현', '인아'])
print(df)
print('\n')

# 데이터 프레임 df를 복제해 변수 df2에 저장. df2의 행 하나 삭제
df2 = df[:]
df2.drop('우현', inplace=True)
print(df2)
print('\n')

# 데이터 프레임 df 복제후 df3에 저장. df3 행 2개 삭제
df3 = df[:]
df3.drop(['우현', '인아'], axis=0, inplace=True)
print(df3)
````

## 예제 1-9 열 삭제

````python
df4 = df.copy()
df4.drop('수학', axis=1, inplace=True)
print(df4)
print('\n')

# 데이터 프레임 df를 복제해 변수 df5에 저장 열 2개 삭제
df5 = df.copy()
df5.drop(['영어', '음악'], axis=1, inplace=True)
print(df5)
````

## 예제 1-9 행 선택

````python
# 행 인덱스로 행 1개 선택
label1 = df.loc['서준']
position1 = df.iloc[0]
print(label1)
print('\n')
print(position1)

# 행 인덱스 2개 선택
label2 = df.loc[['서준', '우현']]
position2 = df.iloc[[0, 1]]
print(label2)
print('\n')
print(position2)

# 행 인덱스 범위로 지정
label3 = df.loc['서준' : '우현']
position3 = df.iloc[0:1]
print(label3)
print('\n')
print(position3)
````

## 예제 1-10 열 선택
#### __열 1개 선택(시리즈 생성) : ```DataFrame 객체['열이름'] 또는 DataFrame 객체.열이름```__
#### __열 n개 선택(데이터 프레임 생성) : ```DataFrame 객체 [[열1, 열2, .....]]```__

````python
# Dataframe() 함수로 데이터 프레임 저장
exam_data = {'이름' : ['서준', '우현', '인아'],
             '수학' : [90, 80, 70],
             '영어' : [98, 89, 95],
             '음악' : [85, 95, 100],
             '체육' : [100, 90, 90]}
df = pd.DataFrame(exam_data)
print(df)
print(type(df))
print('\n')

# 수학 데이터만 선택
math1 = df['수학']
print(math1)
print(type(math1))
print('\n')

# 영어 점수 데이터만 선택
english = df.영어
print(english)
print(type(english))
print('\n')

# 음악, 체육 점수 데이터 선택
music = df[['음악', '체육']]
print(music)
print(type(music))
print('\n')

# 수학 점수만 선택
math2 = df[['수학']]
print(math2)
print(type(math2))
````

## 예제 1-11 원소 선택
#### __인덱스 이름 : ```DataFrame 객체.loc[행 인덱스, 열이름]```__
#### __정수 위치 인덱스 : ```DataFrame 객체.iloc[행번호, 열번호]```__

````python
#이름 열을 새로운 인덱스로 지정후 df 객체에 변경사항 반영
df.set_index('이름', inplace = True)
print(df)
````

````python
# 데이터 프레임 df의 특정 원소 1개 선택
a = df.loc['서준', '음악']
print(a)
b = df.iloc[0, 2]
print(b)
````

````python
# 데이터 프레임 df의 특정 원소 2개 이상 선택
c = df.loc['서준', ['음악', '체육']]
print(c)
d = df.iloc[0, [2, 3]]
print(d)
e = df.loc['서준', '음악' : '체육']
print(e)
f = df.iloc[0, 2:]
print(f)
````

````python
# df 2개 이상의 행과 열에 속하는 원소 2개 선택
g = df.loc[['서준', '우현'], ['음악', '체육']]
print(g)
h = df.iloc[[0, 1], [2, 3]]
print(h)
i = df.loc['서준' : '우현', '음악' : '체육']
print(i)
j = df.iloc[0:2, 2:]
print(j)
````

## 예제 1-14 원소 값 변경
#### __원소 값 변경 : ```DataFrame 객체의 일부분 또는 원소 선택 = 새로운 값```__

````python
exam_data = {'이름' : ['서준', '우현', '인아'],
             '수학' : [90, 80, 70],
             '영어' : [98, 89, 95],
             '음악' : [85, 95, 100],
             '체육' : [100, 90, 90]}
df = pd.DataFrame(exam_data)
print(df)
print(type(df))
print('\n')

# 이름 열을 새로운 인덱스로 지정
df.set_index('이름', inplace=True)
print(df)
print('\n')
````

## 예제 1-15 행, 열 변경
#### __행, 열 변경 : ```DataFrame 객체.transpose() 또는 DataFrame 객체.T```__

````python
exam_data = {'이름' : ['서준', '우현', '인아'],
             '수학' : [90, 80, 70],
             '영어' : [98, 89, 95],
             '음악' : [85, 95, 100],
             '체육' : [100, 90, 90]}
df = pd.DataFrame(exam_data)


# DF를 전치
df = df.transpose()
print(df)
print('\n')

# DF 다시 전치
df = df.T
print(df)
````

## 예제 1-17 새로운 배열로 행 인덱스 재지정
#### __새로운 배열로 행 인덱스를 재지정 : ```DataFrame 객체.reindex(새로운 인덱스 배열)```__

````python
# 딕셔너리 정의
dict_data = {'c0' : [1,2,3], 'c1' : [4, 5, 6], 'c2' : [7,8,8], 'c3' : [10, 11, 12], 'c4' : [13, 14, 15]}

# 딕셔너리를 데이터프레임으로 변환
df = pd.DataFrame(dict_data, index=['r0', 'r1', 'r2'])
print(df)
print('\n')

# 인덱스를 r0, r1, r2, r3, r4로 지정
new_index = ['r0', 'r1', 'r2', 'r3', 'r4']
ndf = df.reindex(new_index)
print(ndf)
print('\n')
````

````python
# reindex로 발생한 NaN 값을 숫자 0으로 채우기
new_index = ['r0', 'r1', 'r2', 'r3', 'r4']
ndf2 = df.reindex(new_index, fill_value=0)
print(ndf2)
````
