## 예제 5-1 누락 데이터 확인

````python
import seaborn as sns

df = sns.load_dataset('titanic')
````

````python
df.info()
````

````python
# deck 열의 NaN 개수 계산
nan_deck = df['deck'].value_counts(dropna=False)
print(nan_deck)
````

* __```innull()``` : 누락 데이터면 True를 반환하고, 유효한 데이터가 존재하면 False 반환__
* __```notnull()``` : 유효 데이터가 존재시 True 반환, 누락 데이터면 False 반환__

````python
# isnull() 메소드로 누락 데이터 찾기
print(df.head().isnull())
````

````python
# notnull() 메소드로 누락 데이터 찾기
print(df.head().notnull())
````

````python
# isnull() 메소드로 누락 데이터 개수 구하기
print(df.head().isnull().sum(axis=0))
````

## 예제 5-2 누락 데이터 제거

````python
import seaborn as sns

# for문으로 각 열의 NaN 개수 계산
missing_df = df.isnull()
for col in missing_df.columns:
    missing_count = missing_df[col].value_counts()

    try:
        print(col, ': ', missing_count[True])
    except:
        print(col, ': ', 0)
````

````python
# NaN 값이 500개 이상인 열을 모두 삭제
df_thresh = df.dropna(axis=1, thresh=500)
print(df_thresh.columns)
````

## 예제 5-3 평균으로 누락 데이터 바꾸기

````python
# age 열의 첫 10개 데이터 출력
print(df['age'].head())
print('\n')

# age 열의 NaN 값을 다른 나이 데이터의 평균으로 변경
mean_age = df['age'].mean(axis=0)
df['age'].fillna(mean_age, inplace=True)

# age 열의 첫 10 개 데이터 출력
print(df['age'].head())
````

````python
# age 열에 나이 데이터가 없는 모든 행 삭제
df_age = df.dropna(subset=['age'], how='any', axis=0)
print(len(df_age))
````
